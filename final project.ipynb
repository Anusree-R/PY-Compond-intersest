```python
#calc using tkinter
import tkinter as tk 
from tkinter import *
root = Tk()
root.configure(background = '#e6e6e6')
root.title("CI Calculator") 
variable1 = tk.StringVar(root)
variable2 = tk.StringVar(root) 
variable1.set("currency")
variable2.set("currency")
# function for clear all
def clear_all():
    principlea.delete(0, END)
    ratea.delete(0, END)
    timea.delete(0, END)
    compounda.delete(0, END)
    Amount1_field.delete(0, tk.END)
    Amount2_field.delete(0, tk.END)
#function for clear
def clear() :
    compounda.delete(0, END)  
#function for CI
def cal_ci():
    principle = int(principlea.get()) 
    rate = float(ratea.get())
    time = int(timea.get()) 
#CI fomulae
    Amount = principle * (pow((1 + rate / 100), time))
    CI = Amount - principle
    compounda.insert(10, CI)
#function for clear1
def clear1() :
    Amount2_field.delete(0, tk.END)
#cuntion for convereter
def RealTimeCurrencyConversion():
    from forex_python.converter import CurrencyRates
    c = CurrencyRates()
    from_currency = variable1.get()
    to_currency = variable2.get()
    new_amt = c.convert(from_currency, to_currency, float(Amount1_field.get()))
    new_amount = float("{:.4f}".format(new_amt))
    Amount2_field.insert(0, str(new_amount))
#This will provide the data for the optionmenu  
CurrenyCode_list = ["INR", "USD", "CAD", "CNY", "DKK", "EUR"]
#It Allows You to Execute Code When the File Runs as a Script
if __name__ == "__main__" :
#the converter line
    label1 = tk.Label(text = "Enter the amount : ", bg = '#e6e6e6') #used for background with the (placeholder)text, background color of #e6e6e6.
label1.grid(row=1, column=0, padx = 30, pady = 30)
 
label2 = tk.Label(text = "From currency : ", bg = '#e6e6e6')
label2.grid(row=2, column=0, padx = 30, pady = 30)
 
label3 = tk.Label(text = "To currency : ", bg = '#e6e6e6')
label3.grid(row=3, column=0, padx = 30, pady = 30)
 
label4 = tk.Label(text = "Converted Amount : ", bg = '#e6e6e6')
label4.grid(row=5, column=0, padx = 30, pady = 30)
 #the option menu
FromCurrency_option = tk.OptionMenu(root, variable1, *CurrenyCode_list)
ToCurrency_option = tk.OptionMenu(root, variable2, *CurrenyCode_list)
 
FromCurrency_option.grid(row=2, column=1)
ToCurrency_option.grid(row=3, column=1)
 
Amount1_field = tk.Entry(root)
Amount1_field.grid(row=1, column=1)
 
Amount2_field = tk.Entry(root)
Amount2_field.grid(row=5, column=1)
 
Label_9 = Button(text="Convert", bg = '#e6e6e6',command=lambda: [clear1(),RealTimeCurrencyConversion()])
Label_9.grid(row=4, column=1)
#the ci calc 
PrincipleAmount = Label(text = "Principle Amount : ", bg = '#e6e6e6')
Rate = Label(text = "Rate : ",bg = '#e6e6e6')
Time = Label(text = "Time : ", bg = '#e6e6e6')
CompoundInterest = Label(text = "Compound Interest : ", bg='#e6e6e6')
#grid on column2 
PrincipleAmount.grid(row = 1, column = 2, padx = 30, pady = 30) 
Rate.grid(row = 2, column = 2, padx = 30, pady = 30) 
Time.grid(row = 3, column = 2, padx = 30, pady = 30)
CompoundInterest.grid(row = 5, column = 2, padx = 30, pady = 30)
    
#entry grid
principlea = Entry(root) 
ratea = Entry(root) 
timea = Entry(root)
compounda = Entry(root)
    
#entry grid properties
principlea.grid(row = 1, column = 3, padx = 30, pady = 30) 
ratea.grid(row = 2, column = 3, padx = 30, pady = 30) 
timea.grid(row = 3, column = 3, padx = 30, pady = 30)
compounda.grid(row = 5, column = 3, padx = 30, pady = 30)
    
#submit button
button1 = Button(text = "Calculate", bg = "#e6e6e6",fg = "black", command =lambda: [clear(), cal_ci()])#used to execute multiple functions in button
#clear allbutton
button2 = Button(text = "Clear all", bg = "#e6e6e6",fg = "black", command = clear_all)

#button properties    
button1.grid(row = 4, column = 3, pady = 10)
button2.grid(row = 6, column = 3, pady = 10)

root.mainloop()
```
